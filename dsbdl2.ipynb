```python
df['Department']=df['Department'].fillna(df['Department'].mode())
```

```python
df['Physics']=df['Physics'].fillna(df['Physics'].mean())
```

```python
df['Mathematics']=df['Mathematics'].fillna(df['Mathematics'].mean())
```

```python
df['AI']=df['AI'].fillna(df['AI'].mean())
```

```python
plt.boxplot(df['Chemistry'])
```

```python
Q1 = df['Chemistry'].quantile(0.25)
Q3 = df['Chemistry'].quantile(0.75)
IQR = Q3-Q1
upper_limit = Q3+1.5*IQR
lower_limit = Q1-1.5*IQR
```

```python
df[(df['Chemistry']>upper_limit) | (df['Chemistry']<lower_limit)]
```

```python
df=df[~((df['Chemistry']>upper_limit) | (df['Chemistry']<lower_limit))]
```
